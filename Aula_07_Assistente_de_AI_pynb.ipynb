 # **Calculadora Simples**

```python
# Solicita o primeiro número ao usuário e converte para decimal (float)
num1 = float(input("Digite o primeiro número: "))

# Solicita o segundo número ao usuário e converte para decimal (float)
num2 = float(input("Digite o segundo número: "))

# Exibe as opções disponíveis para o usuário no console
print("Escolha uma opção: ")
print("1. somar")
print("2. subtrair")

# Captura a escolha do usuário e padroniza para letras minúsculas para evitar erros de comparação
opcao = input("Digite sua escolha (somar/subtrair): ").lower()

# Inicia a estrutura condicional para verificar a operação escolhida
if opcao == "somar":
    # Se a escolha for 'somar', realiza a adição dos dois números
    resultado = num1 + num2
    # Exibe o resultado da soma formatado na tela
    print(f"A soma é {resultado}")

elif opcao == "subtrair":
    # Se a escolha for 'subtrair', realiza a subtração do primeiro pelo segundo
    resultado = num1 - num2
    # Exibe o resultado da subtração formatado na tela
    print(f"A subtração é {resultado}")

else:
    # Caso o usuário digite algo diferente das opções válidas, exibe uma mensagem de erro
    print("Opção inválida. Por favor, escolha 'somar' ou 'subtrair'.")
```
