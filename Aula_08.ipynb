### Organizador Automático de Arquivos
Este script utiliza as bibliotecas `os` e `shutil` para organizar arquivos por extensão.

```python
import os
import shutil

# 1. Definir o diretório a ser organizado
# No Colab, vamos criar uma pasta de exemplo para testar
diretorio_alvo = '/content/meus_arquivos'

# Criar arquivos de teste se a pasta não existir
if not os.path.exists(diretorio_alvo):
    os.makedirs(diretorio_alvo)
    # Criando arquivos vazios para demonstração
    for f in ['documento.pdf', 'foto.jpg', 'relatorio.pdf', 'script.py', 'imagem.png']:
        open(os.path.join(diretorio_alvo, f), 'a').close()
    print(f"Pasta de teste '{diretorio_alvo}' criada com arquivos fictícios.")

def organizar_arquivos(caminho):
    # Listar todos os itens no diretório
    for item in os.listdir(caminho):
        caminho_completo = os.path.join(caminho, item)

        # Pular se for uma pasta
        if os.path.isdir(caminho_completo):
            continue

        # Extrair a extensão do arquivo (ex: 'pdf')
        nome_arquivo, extensao = os.path.splitext(item)
        extensao = extensao.replace('.', '').lower() # Remove o ponto e deixa minúsculo

        # Se o arquivo não tiver extensão, ignorar
        if not extensao:
            continue

        # 2. Criar a pasta de destino se não existir
        pasta_destino = os.path.join(caminho, extensao)
        if not os.path.exists(pasta_destino):
            os.makedirs(pasta_destino)

        # 3. Mover o arquivo
        try:
            shutil.move(caminho_completo, os.path.join(pasta_destino, item))
            print(f"Movido: {item} -> {extensao}/")
        except PermissionError:
            print(f"Erro de permissão ao mover: {item}")
        except Exception as e:
            print(f"Erro ao mover {item}: {e}")

# Executar a função
organizar_arquivos(diretorio_alvo)
print('\nOrganização concluída!')
```

### Consulta de Endereço via API ViaCEP
Este script realiza requisições HTTP para buscar informações de endereços a partir do CEP.

```python
import requests
import json

def consultar_cep(cep):
    # Limpar o CEP (remover espaços e traços)
    cep = str(cep).replace('-', '').replace(' ', '')

    # Validar formato básico (8 dígitos)
    if len(cep) != 8 or not cep.isdigit():
        print(f"Erro: O CEP '{cep}' é inválido. Certifique-se de usar 8 dígitos numéricos.")
        return

    url = f'https://viacep.com.br/ws/{cep}/json/'

    try:
        response = requests.get(url, timeout=10)
        # Levanta exceção para erros HTTP (404, 500, etc)
        response.raise_for_status()

        dados = response.json()

        # A API ViaCEP retorna 'erro': True se o CEP não for encontrado
        if 'erro' in dados:
            print(f"CEP {cep} não encontrado na base de dados.")
        else:
            print("--- Informações do Endereço ---")
            print(f"Logradouro: {dados.get('logradouro', 'N/A')}")
            print(f"Bairro:     {dados.get('bairro', 'N/A')}")
            print(f"Cidade/UF:  {dados.get('localidade', 'N/A')} - {dados.get('uf', 'N/A')}")
            print(f"DDD:        {dados.get('ddd', 'N/A')}")
            print("-------------------------------")

    except requests.exceptions.ConnectionError:
        print("Erro: Não foi possível conectar ao servidor. Verifique sua internet.")
    except requests.exceptions.HTTPError as e:
        print(f"Erro HTTP: {e}")
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")

# Solicitar entrada do usuário
if __name__ == '__main__':
    cep_input = input('Digite o CEP para consulta (ex: 01001000): ')
    consultar_cep(cep_input)
```
